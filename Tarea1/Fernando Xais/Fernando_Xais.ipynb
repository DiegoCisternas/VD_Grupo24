#### 1 Cuales fueron los artistas que mas se repitieron en el top 100 canciones entre 1990-1999

```python
#%pip install plotly.express
#%pip install nbformat
#!python -m pip install nbformat --upgrade
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset_to_use.csv")

top_artistas = df.groupby("Artista")["Apariciones"].sum().sort_values(ascending=False).head(15)
top_artistas = top_artistas[::-1]

# Lollipop chart
plt.figure(figsize=(10, 6))
plt.hlines(y=top_artistas.index, xmin=0, xmax=top_artistas.values, color='skyblue')
plt.plot(top_artistas.values, top_artistas.index, "o", color='darkblue')

plt.title("Top 15 artistas con más apariciones en el top 100 1990-1999")
plt.xlabel("Apariciones")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#### 2 De donde vienen los artistas que componen el top 100 canciones entre 1990-1999

```python

fig = px.sunburst(df, path=['Region', 'Pais', 'Artista'], values=None,color='Region',
  color_discrete_map={
    "North America": "0FB5AE",
    "Europe": "4046CA",
    "Oceania": "00B900"
  },
  width=800,
  height=700

)

fig.update_layout(title_text="Lugar de origen de los artistas que componen el top 100 1990-1999", title_x = 0.5)
fig.show()
```
