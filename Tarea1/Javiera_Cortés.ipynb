```
import matplotlib.pyplot as plt
import squarify

genre_counts = df_1990s['genre'].value_counts().sort_values(ascending=False)

percentages = (genre_counts / genre_counts.sum() * 100).round(1)
labels = [f"{genre}\n({percentage}%)" for genre, percentage in zip(genre_counts.index, percentages)]

sizes = genre_counts.values
normed = squarify.normalize_sizes(sizes, 100, 100)
rects = squarify.squarify(normed, 0, 0, 100, 100)

colors = plt.cm.Set3.colors 
import itertools
color_cycle = list(itertools.islice(itertools.cycle(colors), len(genre_counts)))

fig, ax = plt.subplots(figsize=(10, 10))

for i, rect in enumerate(rects):
    x, y, dx, dy = rect['x'], rect['y'], rect['dx'], rect['dy']
    ax.add_patch(plt.Rectangle((x, y), dx, dy, color=color_cycle[i], alpha=0.8))
    
    if dx > 5 and dy > 5:
        ax.text(
            x + dx/2, y + dy/2,
            labels[i],
            fontsize=10,
            va='center', ha='center'
        )

plt.title('Géneros en 1990', fontsize=16)
plt.axis('off')
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

```
selected_genres = ['pop', 'hip hop', 'rock']
df_1990s_filtered = df_1990s[df_1990s['genre'].isin(selected_genres)]
columns_to_average = ['energy', 'acousticness', 'danceability']

averages_1990s_genres = df_1990s_filtered.groupby('genre')[columns_to_average].mean()
averages_1990s_genres
```

```
selected_genres = ['pop rock', 'r&b', 'soft rock']
df_1990s_filtered = df_1990s[df_1990s['genre'].isin(selected_genres)]
columns_to_average = ['energy', 'acousticness', 'danceability']

averages_1990s_genres2 = df_1990s_filtered.groupby('genre')[columns_to_average].mean()
averages_1990s_genres2
```

```
import matplotlib.pyplot as plt
import numpy as np

labels = averages_1990s_genres.columns.tolist()
genres_group1 = ['pop', 'hip hop', 'rock']  # ordered by saturation
genres_group2 = ['pop rock', 'r&b', 'soft rock']  # ordered by saturation
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

group1_colors = ['#FF5733', '#FF8D72', '#FFC0CB']  # Red, Coral, Light Pink
group2_colors = ['#0000FF', '#3399FF', '#99CCFF']  # Blue, Sky Blue, Light Blue

for genre, color in zip(genres_group1, group1_colors):
    values = averages_1990s_genres.loc[genre].tolist()
    values += values[:1]
    ax.plot(angles, values, label=genre, color=color, linewidth=2)
    ax.fill(angles, values, alpha=0.05, color=color)

for genre, color in zip(genres_group2, group2_colors):
    values = averages_1990s_genres2.loc[genre].tolist()
    values += values[:1]
    ax.plot(angles, values, label=genre, color=color, linewidth=2)
    ax.fill(angles, values, alpha=0.05, color=color)

ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles[:-1]), labels)

plt.legend(loc='upper right', bbox_to_anchor=(1.4, 1.1))
plt.title("Gráfico de radar de promedios de género de la década de 1990")
plt.show()
```
